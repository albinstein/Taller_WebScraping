```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Definicion de Variables
data_libros = []
rating_diccionario = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
```

```python
#URL de la pagina web
url = "https://books.toscrape.com/"
response = requests.get(url)
```

```python
if response.status_code == 200:
  #Se analiza el contenido HTML de la página con BeautifulSoup
  soup = BeautifulSoup(response.content, "html.parser")

  #Encontrar todos los elementos que contienen la información de los libros
  books = soup.find_all('article', class_='product_pod')

  for book in books:
    title = book.h3.a['title']
    price = float(book.find('p', class_='price_color').text[1:])
    availability = book.find('p', class_='instock availability').text.strip()
    rating = rating_diccionario[book.p['class'][1]]
    data_libros.append({
              'title': title,
              'price': price,
              'availability': availability,
              'rating': rating
          })

  print(f"Solicitud exitosa, se han encontrado {len(data_libros)} articulos")
  df = pd.DataFrame(data_libros)

else:
  print("Error en la solicitud")
```

```python
#Dataframe con todos los libros que se extrajo
df
```

```python
# Obtener el libro con el mayor rating
book_mayor_rating = df[df['rating'] == df['rating'].max()]
book_mayor_rating
```

```python
# Obtener el libro con el mayor precio
book_mas_caro = df[df['price'] == df['price'].max()]
book_mas_caro
```

```python
# Calcular el precio promedio de los libros
precio_prom = df['price'].mean()
precio_prom
```

```python
# Obtener el libro con el título más largo
book_con_titulo_mas_largo = df[df['title'].str.len() == df['title'].str.len().max()]
book_con_titulo_mas_largo
```
